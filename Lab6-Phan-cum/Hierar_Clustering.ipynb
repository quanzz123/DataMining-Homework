- Mục tiêu:

    + Áp dụng mô hình để giải quyết bài toán thực tế: Phân nhóm hoa Iris.

- Dữ liệu:
    + Dataset 'Iris.csv' có 150 mẫu, trong đó mỗi mẫu là một quan sát chi tiết về một cá thể hoa Iris.
    + Có 4 thuộc tính, trong đó mỗi thuộc tính là một đặc trưng hình thái học (chiều dài, chiều rộng của cánh và đài hoa) của mỗi bông hoa.
- Mô hình hóa bài toán: Bài toán phân cụm
    + Đầu vào: N vector D chiều, tương ứng với N điểm dữ liệu với D thuộc tính trong không gian.

### 2. Các bước thực hiện

2.1. Import thư viện cần thiết

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA 
from sklearn.cluster import AgglomerativeClustering 
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score 
import scipy.cluster.hierarchy as shc 
```

2.2. Hiểu và tiền xử lý dữ liệu

```python
data = pd.read_csv('Iris.csv')
data.info()
```

```python
# Kiểm tra missing value
data.isnull().sum()
```

```python
# Xóa cột Id và Species
data = data.drop(['Id', 'Species'], axis = 1)
data.info()
```

```python
# Chuẩn hóa dữ liệu (Normalization)
scaler = StandardScaler() 
X_scaled = scaler.fit_transform(data) 
X_scaled = pd.DataFrame(X_scaled) 
X_scaled
```

2.2. Giảm chiều dữ liệu và trực quan hóa phân bố của các mẫu dữ liệu bằng biểu đồ phân tán (Scatter Plot)

```python
# Giảm chiều dữ liệu bằng PCA (Principal component analysis) với n_components = 2
pca = PCA(n_components = 2) 
X_principal = pca.fit_transform(X_scaled) 
X_principal = pd.DataFrame(X_principal) 
X_principal.columns = ['P1', 'P2']
```

```python
# Trực quan hóa các điểm dữ liệu trên đồ thị scatter plot
plt.figure(figsize=(9, 6))
plt.scatter(X_principal['P1'], X_principal['P2'], lw=0, s=40)
plt.xlabel('P1')
plt.ylabel('P2')
plt.title('Phân bố các điểm dữ liệu')
```

2.3. Trực quan hóa quá trình phân cấp cụm dữ liệu bằng dendrogram

```python
plt.figure(figsize =(9, 6)) 
plt.title('Dendrogram') 
Dendrogram = shc.dendrogram((shc.linkage(X_principal, method ='ward'))) 
```

2.4. Phân cụm phân cấp theo chiến lược hợp nhất sử dụng AgglomerativeClustering

```python
ac2 = AgglomerativeClustering(n_clusters = 2) 
labels_ac2 = ac2.fit_predict(X_principal)
# Trực quan hóa kết quả phân cụm
plt.figure(figsize =(6, 6)) 
plt.scatter(X_principal['P1'], X_principal['P2'],  c = labels_ac2, cmap ='rainbow') 
plt.show() 
```

- Đánh giá chất lượng phân cụm bằng chỉ số silhouette_score (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html, https://medium.com/@MrBam44/how-to-evaluate-the-performance-of-clustering-algorithms-3ba29cad8c03)

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import os
```

```python
path = r"Z:\DOCUMENTS\KPDL\DataMining-Homework\Lab6-Phan-cum\shopping_data.csv"
```

```python
df = pd.read_csv(path)
print("---Thông tin bộ dự liệu---")
print(df.info())
print("\n---5 dong đầu tiên ---")
print(df.head())
```

```python
# Cột 3: Annual Income (k$), Cột 4: Spending Score (1-100)
data = df.iloc[:, [3, 4]].values
data
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], color='blue', edgecolors='k', alpha=0.7)
plt.title("Phân bố dữ liệu: Thu nhập hàng năm vs Điểm chi tiêu")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 3.3. Trực quan hóa quá trình phân cấp bằng Dendrogram 
plt.figure(figsize=(12, 7))
plt.title("Dendrogram - Shopping Data")
# Sử dụng phương pháp 'ward' để giảm thiểu phương sai trong cụm
dend = shc.dendrogram(shc.linkage(data, method='ward'))
# plt.axhline(y=150, color='r', linestyle='--')  
plt.xlabel("Số thứ tự điểm dữ liệu")
plt.ylabel("Khoảng cách Euclidean")
plt.show()
```

```python
# --- 3.4. Xây dựng mô hình AgglomerativeClustering với n_clusters = 4 ---
model = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
labels = model.fit_predict(data)
```

```python
# Trực quan hóa kết quả phân cụm
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='rainbow', edgecolors='k', s=70)
plt.title("Kết quả phân cụm Shopping Data (n_clusters = 4)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
